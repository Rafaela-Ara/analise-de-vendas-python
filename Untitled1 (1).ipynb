```python
import pandas as pd
```

```python
df = pd.read_csv('dados_vendas_ficticios.csv')
```

```python
df['Data do Pedido'] = pd.to_datetime(df['Data do Pedido'], errors='coerce')
df['Mês'] = df['Data do Pedido'].dt.to_period('M')
```

```python
df.columns
```

```python
# Verificar se a coluna 'Data do Pedido' foi convertida para datetime corretamente
df['Data do Pedido'] = pd.to_datetime(df['Data do Pedido'], errors='coerce')

# Criar a coluna 'Mês'
df['Mês'] = df['Data do Pedido'].dt.to_period('M')
```

```python
import pandas as pd
import matplotlib.pyplot as plt # Import the matplotlib library

# ... (rest of your code) ...

# Agrupar por mês e somar as receitas
vendas_mensais = df.groupby('Mês')['Receita'].sum()

# Plotar gráfico de barras
vendas_mensais.plot(kind='bar', figsize=(10, 5), color='skyblue', title='Receita Mensal')
plt.ylabel('Receita (R$)') # Now plt is recognized
plt.xlabel('Mês')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Agrupar por produto e somar a quantidade vendida
produtos_mais_vendidos = df.groupby('Produto')['Quantidade'].sum().sort_values(ascending=False).head(10)

# Plotar gráfico de barras horizontais
produtos_mais_vendidos.plot(kind='barh', figsize=(8, 5), color='orange', title='Top 10 Produtos Mais Vendidos')
plt.xlabel('Quantidade Vendida')
plt.tight_layout()
plt.show()
```

```python
# Agrupar por região e somar as receitas
regiao_receita = df.groupby('Região')['Receita'].sum().sort_values(ascending=False)

# Plotar gráfico de pizza
regiao_receita.plot(kind='pie', autopct='%1.1f%%', figsize=(6, 6), title='Distribuição de Receita por Região')
plt.ylabel('')
plt.show()
```

```python
# Calcular o crescimento percentual da receita mês a mês
vendas_mensais = df.groupby('Mês')['Receita'].sum()

# Calcular o crescimento percentual
crescimento_vendas = vendas_mensais.pct_change() * 100

# Exibir o crescimento das vendas
print(crescimento_vendas)
```

```python
# Plotar gráfico de linha para o crescimento das vendas
crescimento_vendas.plot(kind='line', figsize=(10, 5), color='green', title='Crescimento Percentual das Vendas')
plt.ylabel('Crescimento (%)')
plt.xlabel('Mês')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
